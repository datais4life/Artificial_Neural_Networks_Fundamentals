```python
# Import needed libraries, cv2 to read the image and matplotlib to plot the image.
%matplotlib inline
import cv2, matplotlib.pyplot as plt

# Read the image
img = cv2.imread(r"C:\Users\jilla\Documents\Projects\PyTorch\data\Hemanvi.jpeg")
```

```python
# Crop the image between the 50th and 250th rows and the 40th and 240th columns.
img = img[50:250,40:240]

# Convert the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Show the image
plt.imshow(img_gray, cmap='gray')
```

```python
# Convert the image to a 25 x 25 array and plot it.
img_gray_small = cv2.resize(img_gray,(25,25))
plt.imshow(img_gray_small, cmap='gray')
```
